```python
# replace non-numeric fields to numerics fields
lungcapdata.Smoke.replace( {"no": 1, "yes": 0 },  inplace=True)
```

```python
lungcapdata_x=lungcapdata.iloc[:,1:6]
lungcapdata_y=lungcapdata.iloc[:,0]
```

```python
lungcapdata_x.shape
```

```python
lungcapdata_y.shape
```

```python
lungcapdata_x_train,lungcapdata_x_test,lungcapdata_y_train,lungcapdata_y_test=train_test_split(lungcapdata_x,lungcapdata_y,test_size=0.2)
```

```python
lungcapdata_x_train.shape
```

```python
lungcapdata_x_test.shape
```

```python
#create the model
mymodel_model=LinearRegression()

#train the model
mymodel_model.fit(lungcapdata_x_train,lungcapdata_y_train)
```

```python
#test the model
pred_y=mymodel_model.predict(lungcapdata_x_test)
pred_y
```

```python
#check shape of the predicted vestor should be same as test y values
print(pred_y.shape)
print(lungcapdata_y_test.shape)
```

```python
#check the values of model parameters/weights/coefficients
mymodel_model.coef_ # y =w0+w1.x1+w2.x2+w3.x3+w4.x4+w5.x5
```

```python
#check the corresponding x names
lungcapdata_x_train.columns
```

```python
mymodel_model.intercept_
```

```python
# check r-square metric to find out the performance of the trained model, higher the better
mymodel_square=mymodel_model.score(lungcapdata_x_train,lungcapdata_y_train)
print(mymodel_square)
```

```python
#Adjusted R Square  
# k= number of columns in train set
# n= number of rows(samples)in train set

n=lungcapdata_x_train.shape[1]
m=lungcapdata_x_train.shape[0]
print(m,n)

#formula

adj_mymodel_square=1 - ((1-mymodel_square)*(m-1))/(m-n-1)
print(adj_mymodel_square)
```

```python
#MAE on test data
lungcapdata_mae=metrics.mean_absolute_error(lungcapdata_y_test,pred_y) #actual y values - predicted y values
print(lungcapdata_mae)
```

```python
#MSE on test data
lungcapdata_mse=metrics.mean_squared_error(lungcapdata_y_test,pred_y)
print(lungcapdata_mse)
```

```python
#MSE on test data
lungcapdata_mse=metrics.mean_squared_error(lungcapdata_y_test,pred_y)
print(lungcapdata_mse)
```

```python
#MAPE (Mean Absolute Percentage Error) on test data
i=((lungcapdata_y_test-pred_y)/lungcapdata_y_test)*100
i_abs=np.abs(i)

lungcapdata_mape=np.mean(i_abs)

print('mape:',lungcapdata_mape)
print('accuracy:',100-lungcapdata_mape)
```
